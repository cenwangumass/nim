```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
x = np.load("train.npy")
```

```python
diff = np.diff(x, axis=1)
m = diff.mean()
std = diff.std()
```

```python
y = np.random.randn(200) * std + m
y = np.cumsum(y)
```

```python
plt.plot(y)
```

```python
model = SARIMAX(y, order=(0, 1, 1), seasonal_order=(1, 0, 0, 12))
result = model.fit()
result.summary()
result.llf
```
