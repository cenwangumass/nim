```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.load("train.npy")
x = x[0]

y = np.diff(x)
```

```python
plt.plot(x)
plt.xlabel("Timestep")
plt.ylabel("Value")
plt.savefig("plots/sarima-sample.png", dpi=200)
```

```python
plt.plot(y)
plt.xlabel("Timestep")
plt.ylabel("Value")
plt.savefig("plots/sarima-diff-sample.png", dpi=200)
```
